```python
def RMSE(y_true,y_pred):
  output = np.sqrt(np.mean((y_true-y_pred)**2))
  return output
```

```python
def RMSLE(y_true,y_pred):
  output = np.sqrt(np.mean((np.log(y_true+1)-np.log(y_pred+1))**2))
  return output
```
